```python
# Data of the all the listed companies on a perticular day

import datetime

# Get the current date
current_date = datetime.date.today()

# Calculate yesterday's date
yesterday_date = current_date - datetime.timedelta(days=1)

# Format yesterday's date as 'dd-mm-yyyy'
formatted_date = yesterday_date.strftime('%d-%m-%Y')

# Call the function with yesterday's date
data = capital_market.bhav_copy_equities(formatted_date)
```

Data Variables

- **SYMBOL**: The unique identifier for each stock.
- **OPEN**: The price at which the stock opened for trading on that day.
- **HIGH**: The highest price the stock reached during trading on that day.
- **LOW**: The lowest price the stock reached during trading on that day.
- **CLOSE**: The price at which the stock closed for trading on that day.
- **PREVCLOSE**: The closing price of the stock on the previous trading day.
- **TOTTRDQTY**: The total number of shares traded on that day.
- **TOTTRDVAL**: The total value of shares traded on that day (in rupees).
- **TIMESTAMP**: The date and time of the data.

* Minimum quantity traded: 0.5% of the total number of equity shares of the company listed on the stock exchanges.

* Disclosed within 1 working day of the transaction.
* Reported by the stock exchanges on a daily basis.


```python
capital_market.bulk_deal_data(period='6M')
```

* Minimum quantity traded: 0.5% of the total number of equity shares of the company listed on the stock exchanges.
* Minimum value traded: ₹10 crore.
* Disclosed within 1 working day of the transaction.
* Reported by the stock exchanges on a daily basis.

**Key Differences:**

* **Minimum Value:** Bulk deals have a higher minimum value of ₹10 crore, while block deals have a lower minimum value of ₹5 crore.
* **Purpose:** Bulk deals are typically done by institutional investors for investment purposes, while block deals can be done for a variety of reasons, such as divestment, acquisition, or change in shareholding pattern.
